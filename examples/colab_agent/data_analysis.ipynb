# Data Analysis Agent

This notebook is an example Colab agent for AX. It reads a query from
the `input` variable (set by the AX colab agent before `%run`), generates
synthetic data, performs analysis, and produces a chart.

## Usage via AX

```yaml
registry:
  colab_agents:
    - id: "data-analysis"
      name: "Data Analysis"
      description: "Analyzes data using a Colab notebook on Google Drive."
      remote_file: "/content/drive/MyDrive/notebooks/data_analysis.ipynb"
      drive_mount_path: "/content/drive"
      input_flag: "query"
      output_image: "./examples/colab_agent/chart.png"
```

Then run:
```
ax exec --agent data-analysis --input "Show monthly revenue trend for 2024"
```

````python
# The `query` variable is set by the AX colab agent before %run.
# When running standalone, set it manually.
query = "Show monthly revenue trend for 2024"
try:
    query
except NameError:
    query = "Show monthly revenue trend for 2024"
````

````python
print(f"Query: {input}", flush=True)
print("", flush=True)
````

````python
import time
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
````

````python
# Step 1: Generate synthetic monthly data.
print("[1/3] Generating data...", flush=True)
time.sleep(1)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

np.random.seed(42)
base = np.linspace(100, 180, 12)
noise = np.random.normal(0, 10, 12)
revenue = base + noise
revenue = np.maximum(revenue, 0)

print(f"  Generated 12 months of revenue data", flush=True)
print(f"  Range: ${revenue.min():.0f}K - ${revenue.max():.0f}K", flush=True)
````

````python
# Step 2: Compute statistics.
print("[2/3] Analyzing...", flush=True)
time.sleep(2)

avg = np.mean(revenue)
growth = ((revenue[-1] - revenue[0]) / revenue[0]) * 100
best_month = months[np.argmax(revenue)]
worst_month = months[np.argmin(revenue)]

print(f"  Average monthly revenue: ${avg:.0f}K", flush=True)
print(f"  Year-over-year growth: {growth:.1f}%", flush=True)
print(f"  Best month: {best_month} (${revenue.max():.0f}K)", flush=True)
print(f"  Worst month: {worst_month} (${revenue.min():.0f}K)", flush=True)
````

````python
# Step 3: Generate chart.
print("[3/3] Generating chart...", flush=True)
time.sleep(2)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(months, revenue, color='#2563eb', alpha=0.8)
ax.plot(months, revenue, color='#1d4ed8', linewidth=2, marker='o', markersize=6)
ax.axhline(y=avg, color='#dc2626', linestyle='--', linewidth=1, label=f'Average: ${avg:.0f}K')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Revenue ($K)', fontsize=12)
ax.set_title('Monthly Revenue Trend - 2024', fontsize=14)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)

output_path = './chart.png'
fig.savefig(output_path, dpi=150, bbox_inches='tight')
plt.close(fig)

print(f"  Chart saved to {output_path}", flush=True)
print("", flush=True)
print("Done.", flush=True)
````
